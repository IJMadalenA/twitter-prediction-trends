```python
import numpy as np
import pandas as pd
```

```python
"""Crea los csv de trends por mes"""
```

```python
#cargamos nuestra lista de trends 
df = pd.read_csv("trends.csv", sep = ";")
df
```

```python
#filtramos solo por españa
df1=df[df['location']=='Spain']
df1
```

```python
#añadimos columnas de día y mes
df1["day"]= df1.date.apply(lambda x : x[8:10])
df1["month"] = df1.date.apply(lambda x : x[5:7])
```

```python
df1
```

```python
#Al hacer scrapping, se han colado los municipios, por lo que vamos a eliminar esas filas
```

```python
df1=df1[~(df1["trend"].isin(["Barcelona","Bilbao","Las Palmas","Madrid","Malaga","Murcia","Palma","Seville",
"Valencia","Zaragoza"]))]
```

```python
df1
```

# ENERO
El mes de enero comienza en el día 07 en nuestra búsqueda. Filtraremos los días anteriores

```python
df_enero=df1[~(df1["day"].isin(["00","01","02","03","04","05","06"])) & (df1["month"]=="01")]
df_enero.to_csv("trends_enero.csv", sep = ";", index = False)
df_enero
```

# FEBRERO

```python
df_febrero=df1[~(df1["day"].isin(["00"])) & (df1["month"]=="02")]
df_febrero.to_csv("trends_febrero.csv", sep = ";", index = False)
df_febrero
```

# MARZO

```python
df_marzo=df1[~(df1["day"].isin(["00"])) & (df1["month"]=="03")]
df_marzo.to_csv("trends_marzo.csv", sep = ";", index = False)
df_marzo
```

# ABRIL

```python
df_abril=df1[~(df1["day"].isin(["00"])) & (df1["month"]=="04")]
df_abril.to_csv("trends_abril.csv", sep = ";", index = False)
df_abril
```

```python
import numpy as np
import pandas as pd

from time import sleep

from bs4 import BeautifulSoup
import requests
```

```python
jan = [["2020", "01", "0" + str(num)] if len(str(num)) < 2 else ["2020", "01", str(num)] for num in range(1, 32)]

feb = [["2020", "02", "0" + str(num)] if len(str(num)) < 2 else ["2020", "02", str(num)] for num in range(1, 30)]

mar = [["2020", "03", "0" + str(num)] if len(str(num)) < 2 else ["2020", "03", str(num)] for num in range(1, 32)]

apr = [["2020", "04", "0" + str(num)] if len(str(num)) < 2 else ["2020", "04", str(num)] for num in range(1, 31)]

dates = []

for month in [jan, feb, mar, apr]:
    dates.extend(month)
    
#dates
```

```python
lugares_codigos = {"Spain"      : 23424950,
                   "Barcelona"  : 753692,
                   "Bilbao"     : 754542,
                   "Las Palmas" : 764814,
                   "Madrid"     : 766273,
                   "Malaga"     : 766356,
                   "Murcia"     : 768026,
                   "Palma"      : 769293,
                   "Sevilla"    : 774508,
                   "Valencia"   : 776688,
                   "Zaragoza"   : 779063}
```

```python
all_trends = list()
for lugar, codigo in lugares_codigos.items():
    for date in dates:
        url = "https://trendogate.com/placebydate/{}/{}-{}-{}".format(codigo, date[0], date[1], date[2])
        
        response = requests.get(url)
        print(response)
        soup = BeautifulSoup(response.text, "html.parser")

        trends = [s.text.strip() for s in soup.find_all("li", class_ = "list-group-item")]
        
        for trend in trends:
            all_trends.append([lugar, "{}-{}-{}".format(date[0], date[1], date[2]), trend] )
            
        sleep(10)
```

```python
df = pd.DataFrame(all_trends, columns = ["location", "date", "trend"])
```

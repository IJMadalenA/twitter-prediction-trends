Esta lista de trends incluye, además de Spain, otras comunidades autónomas. Con el fin de agilizar el proceso, filtraremos solo por Spain.

```python
#filtramos solo por españa
df1=df[df['location']=='Spain']
df1
```

Añadimos columnas de día y mes

```python
df1['day']= df1.date.apply(lambda x : x[8:10])
df1["month"] = df1.date.apply(lambda x : x[5:7])
```

Al hacer scrapping, se han colado los nombres de las comunidades como si fueran una tendencia más, por lo que vamos a eliminar esas filas.

```python
df1=df1[~(df1['trend'].isin(['Barcelona','Bilbao','Las Palmas','Madrid','Malaga','Murcia','Palma','Seville',
'Valencia','Zaragoza']))]
df1
```

El mes de enero comienza en el día 07 en nuestra búsqueda. Así que además de filtrarlo por mes, también habrá que filtrarlo por días.

```python
df_enero=df1[~(df1['day'].isin(['00','01','02','03','04','05','06'])) & (df1['month']=='01')].reset_index()
```

```python
trends = list(df_enero.trend)
for trend in trends:
    c = twint.Config()
    c.Search = trend
    c.Lang = "es"
    c.Since = "2020-01-07"
    c.Until = "2020-05-01"
    c.Store_csv = True
    c.Output = "{}.csv".format("_".join(trend.strip("#").split(" ")))
    c.Limit = 20_000
    twint.run.Search(c)
```

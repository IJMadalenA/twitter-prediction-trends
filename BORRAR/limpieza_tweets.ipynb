Eliminamos letras con tilde y sustituirlas por la misma sin tilde

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.lower()
        s = s.replace(a, b)
    return s
```

```python
df['tweet']=df['tweet'].apply(normalize)
```

Eliminamos los hashtags y menciones dentro del tweet

```python
def remove(tweet):
    tweet = re.sub( r"#[A-z0-9_]+", '', tweet)
    tweet = re.sub( r"@[A-zA-z0-9_]+", '', tweet)
    return tweet
```

```python
df['tweet']=df['tweet'].apply(remove)
```

Eliminamos los emojis dentro del tweet

```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
df['tweet']=df['tweet'].apply(deEmojify)
```

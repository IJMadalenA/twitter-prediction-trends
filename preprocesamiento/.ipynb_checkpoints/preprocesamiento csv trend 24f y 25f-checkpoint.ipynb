```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Daniel/Desktop/csv/limpio_tendencias_24-25.csv")
df.drop([df.columns[0], "hours"], axis = 1, inplace = True)
```

```python
df_24, df_25 = df[df.day == 24], df[df.day == 25]
```

```python
df_24
```

```python
df_24["hour"] = df_24.time.apply(lambda x : str(x)[:2])
df_24["minute"] = df_24.time.apply(lambda x : str(x)[3:5])
df_24["second"] = df_24.time.apply(lambda x : str(x)[6:])
```

```python
df_24
```

```python
df_24["mentions_count"] = [len(mention.split(",")) if type(mention) == str else 0 for mention in df_24.mentions]

df_24["reply_count"] = [len(reply.split(",")) if type(reply) == str else 0 for reply in df_24.reply_to]

df_24["hashtags_count"] =  [len(hashtag.split(",")) if type(hashtag) == str else 0 for hashtag in df_24.hashtags]
```

```python
df_summary = pd.DataFrame(df_24.trend.value_counts()).reset_index()
df_summary.columns = ["trend", "total_count"]
```

```python
df_summary
```

```python
df_summary["total_hashtags"] = [df_24[df_24.trend == trend].hashtags_count.sum() for trend in df_summary.trend]

df_summary["total_mentions"] = [df_24[df_24.trend == trend].mentions_count.sum() for trend in df_summary.trend]

df_summary["total_reply"] = [df_24[df_24.trend == trend].reply_count.sum() for trend in df_summary.trend]

df_summary["retweets_count"] = [df_24[df_24.trend == trend].retweets_count.sum() for trend in df_summary.trend]

df_summary["url_count"] = [df_24[df_24.trend == trend].urls.sum() for trend in df_summary.trend]

df_summary["photo_count"] = [df_24[df_24.trend == trend].photos.sum() for trend in df_summary.trend]

df_summary["likes_count"] = [df_24[df_24.trend == trend].likes_count.sum() for trend in df_summary.trend]
```

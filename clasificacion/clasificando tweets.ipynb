```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Daniel/Documents/GitHub/SuperProyecto/elmundo2.csv")
```

```python
# 0 negativo
# 1 neutro
# 2 positivo
# 3 compuesto
```

```python
clasificacion = []
for tweet in df.tweet.iloc[:2000]:
    print(tweet)
    clasificacion.append(input())
```

```python
len(clasificacion)
```

```python
df_clf = pd.DataFrame(clasificacion, columns = ["Daniel"])
```

```python
df_clf.loc[list(df_clf[df_clf.Daniel == ""].index), "Daniel"] = [0, 1, 0, 1, 2, 3, 0, 0, 0, 1]
```

```python
def convert(num):
    if num == "10" or num == 10:
        return 0
    elif num == "31" or num == 31:
        return 1
    return int(num)
```

```python
df_clf.Daniel = df_clf.Daniel.apply(convert)
```

```python
df_clf.to_csv("daniel_clasificacion.csv", sep = ";", index = False)
```

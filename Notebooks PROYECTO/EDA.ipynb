# CLUSTERING

**KMEANS**

```python
from sklearn.cluster import KMeans 
from scipy.spatial.distance import cdist 
```

```python
X = df.iloc[:, 1:].values
```

```python
from sklearn.preprocessing import MinMaxScaler
X = MinMaxScaler().fit_transform(X)
X.shape
```

INERCIAS

```python
inercias = [] 
for k in range(1, 10): 
    kmeans = KMeans(k)
    kmeans.fit(X)     
    inercias.append(kmeans.inertia_) 
inercias
plt.plot(range(1, 10), inercias, "bx-") 
plt.xlabel("Ks") 
plt.ylabel("Inercia") 
plt.show() #buscar el codo, es decir, donde var√≠a la inercia
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
kmeans.labels_
```

```python
df.sort_values("total_count", ascending = False)
```

```python
sns.scatterplot(df["total_count"].values, df["total_hashtags"].values, hue = kmeans.labels_, palette = "Set2")
plt.show()
```

```python
sns.scatterplot(df["total_count"].values, df["total_interaction"].values, hue = kmeans.labels_, palette = "Set2")
plt.show()
```

**HIERARCHICAL**
AGGLOMERATIVE

```python
from sklearn.cluster import AgglomerativeClustering 
from sklearn.metrics.cluster import homogeneity_score
```

```python
linkage = ["ward", "complete", "average", "single"]
for link in linkage:
    agglom = AgglomerativeClustering(n_clusters = 3, linkage = link)
    agglom.fit(X)
    print("Linkage tipo", link)
    print(agglom.labels_)
    print("****")
    print(homogeneity_score(agglom.labels_, kmeans.labels_))
    print("*****")
```

**HIERARCHICAL**
DENDROGRAMA

```python
from scipy.spatial import distance_matrix 
from scipy.cluster import hierarchy 
import pylab
```

```python
dist_matrix = distance_matrix(X,X)
```

```python
link=["single", "complete", "average", "weighted", "centroid", "median", "ward"]
for l in link:
    Z = hierarchy.linkage(dist_matrix, l)
    pylab.figure(figsize=(10, 10))
    pylab.title(l)
    dendro = hierarchy.dendrogram(Z)
```

DISTANCIAS

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors = 2)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
distances = np.sort(distances, axis = 0)
distances = distances[:, 1]
plt.plot(distances)
```

```python
epsilon = 0.8
minimumSamples = 7
db = DBSCAN(eps = epsilon, min_samples = minimumSamples).fit(X)
labels = db.labels_
labels
```
